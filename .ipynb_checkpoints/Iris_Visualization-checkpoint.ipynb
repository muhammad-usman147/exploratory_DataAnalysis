```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
file = 'iris.csv'
data = pd.read_csv(file,index_col='Id')
```

```python
data.head()
```

```python
"Total Unique Iris are{}".format(data.Species.unique())
```

### Histogram

Creating a histogram to see the petal length varies in iris flowers

- With Kernel Density Estimation

```python
sns.distplot(a=data['Petal Length (cm)'])
```

- Without Kernel Density Estimation

```python
sns.distplot(a=data['Petal Length (cm)'],kde=False)
```

#### Density Plot

Plotting KDE(Kernel Density Estimation), to understand basic of KDE you can refer it as a smoothed Histogram

```python
fig ,axes= plt.subplots(1,2,figsize=(10,3))
sns.kdeplot(data=data['Petal Length (cm)'],ax=axes[0]).set_title('Without Shade')
sns.kdeplot(data=data['Petal Length (cm)'],shade=True,ax=axes[1],).set_title("Withj shade")
```

### 2 KDE Plots

```python
sns.jointplot(x=data['Petal Length (cm)'],y=data['Sepal Width (cm)'])
sns.jointplot(x=data['Petal Length (cm)'],y=data['Sepal Width (cm)'],kind='hex')
sns.jointplot(x=data['Petal Length (cm)'],y=data['Sepal Width (cm)'],kind='kde')
```

In the plot below, the color-coding shows us how likely we are to see different combinations of sepal width and petal length, where darker parts of the figure are more likely.

Note that in addition to the 2D KDE plot in the center,

- the curve at the top of the figure is a KDE plot for the data on the x-axis (in this case, iris_data['Petal Length (cm)']), and

### Color Coded Plots

For the next part of the tutorial, we'll create plots to understand differences between the species. To accomplish this, we begin by breaking the dataset into three separate files, with one for each species.

```python
iris_set_filepath = 'iris_setosa.csv'
iris_vers_filepath = 'iris_versicolor.csv'
iris_vir_filepath ='iris_virginica.csv'
```

```python
iris_set = pd.read_csv(iris_set_filepath)
iris_vers = pd.read_csv(iris_vers_filepath)
iris_vir = pd.read_csv(iris_vir_filepath)
```

```python
iris_vir.head()
```

```python
sns.distplot(a=iris_set['Petal Length (cm)'],label = 'Iris Setosa',kde=False)
sns.distplot(a=iris_vers['Petal Length (cm)'], label="Iris Versicolor",kde=False)
sns.distplot(a=iris_vir['Petal Length (cm)'], label = 'Iris Virginica',kde=False)

plt.title("Histogram of Petal Lengths, by Species")
plt.legend()
```

```python
sns.distplot(a=iris_set['Petal Length (cm)'],label = 'Iris Setosa')
sns.distplot(a=iris_vers['Petal Length (cm)'], label="Iris Versicolor")
sns.distplot(a=iris_vir['Petal Length (cm)'], label = 'Iris Virginica')

plt.title("Histogram of Petal Lengths, by Species")
plt.legend()
```

```python
sns.kdeplot(data=iris_set['Petal Length (cm)'],label = 'Iris Setosa')
sns.kdeplot(data=iris_vers['Petal Length (cm)'], label="Iris Versicolor")
sns.kdeplot(data=iris_vir['Petal Length (cm)'], label = 'Iris Virginica')

plt.title("Histogram of Petal Lengths, by Species")
plt.legend()
```

```python
sns.kdeplot(data=iris_set['Petal Length (cm)'],label = 'Iris Setosa',shade=True)
sns.kdeplot(data=iris_vers['Petal Length (cm)'], label="Iris Versicolor",shade=True)
sns.kdeplot(data=iris_vir['Petal Length (cm)'], label = 'Iris Virginica',shade=True)

plt.title("Histogram of Petal Lengths, by Species")
plt.legend()
```

One interesting pattern that can be seen in plots is that the plants seem to belong to one of two groups, where Iris versicolor and Iris virginica seem to have similar values for petal length, while Iris setosa belongs in a category all by itself.

In fact, according to this dataset, we might even be able to classify any iris plant as Iris setosa (as opposed to Iris versicolor or Iris virginica) just by looking at the petal length: if the petal length of an iris flower is less than 2 cm, it's most likely to be Iris setosa!
